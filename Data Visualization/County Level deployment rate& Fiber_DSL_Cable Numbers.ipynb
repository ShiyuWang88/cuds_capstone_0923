# Fiber/DSL/Cable Nums

## 2021

```python
fiber2021="""
SELECT StateAbbr as State,
count(*) as Nums
FROM `broadband-data.fcc_form_477.fbd_us_without_satellite_jun2021_v1` 
where TechCode = 50 and Consumer=1
group by State
order by Nums DESC
"""
dsl2021="""
SELECT StateAbbr as State,
count(*) as Nums
FROM `broadband-data.fcc_form_477.fbd_us_without_satellite_jun2021_v1` 
where TechCode in (10,11,12,20) and Consumer = 1
group by State
order by Nums DESC
"""

cable2021="""
SELECT StateAbbr as State,
count(*) as Nums
FROM `broadband-data.fcc_form_477.fbd_us_without_satellite_jun2021_v1` 
where TechCode in (40,41,42,43) and Consumer =1
group by State
order by Nums DESC
"""

fiber2021df = client.query(fiber2021).to_dataframe()
dsl2021df = client.query(dsl2021).to_dataframe()
cable2021df = client.query(cable2021).to_dataframe()
```

```python
#fiber 2021
plt.figure(figsize=(15,8))
sns.barplot(x='State',y='Nums',data=fiber2021df)
plt.xticks(rotation = 45)
plt.xlabel('State',fontsize=20)
plt.ylabel('Counts',fontsize=20)
plt.title('Census Block: Numbers of Fiber 2021',fontsize=30)
plt.tight_layout()
plt.show()
plt.savefig('Census Block: Numbers of Fiber 2021.png')
```

```python
#DSL 2021
plt.figure(figsize=(15,8))
sns.barplot(x='State',y='Nums',data=dsl2021df)
plt.xticks(rotation = 45)
plt.xlabel('State',fontsize=20)
plt.ylabel('Counts',fontsize=20)
plt.title('Census Block: Numbers of DSL 2021',fontsize=30)
plt.tight_layout()
plt.show()
plt.savefig('Census Block: Numbers of DSL 2021.png')
```

```python
#Cable 2021
plt.figure(figsize=(15,8))
sns.barplot(x='State',y='Nums',data=cable2021df)
plt.xticks(rotation = 45)
plt.xlabel('State',fontsize=20)
plt.ylabel('Counts',fontsize=20)
plt.title('Census Block: Numbers of Cable 2021',fontsize=30)
plt.tight_layout()
plt.show()
plt.savefig('Census Block: Numbers of Cable 2021.png')
```

# County Level Deployment Rate

## 2021

### Top 30

```python
# TOP 30
top="""
with cal as(
  SELECT 
Countycode_2 as X,
IFNULL(BroadbandNum/Nums*100,0) as Percentage
FROM `broadband-data.CountyInfo.2021_countyinfo` 
order by Percentage DESC)

select X,
cal.Percentage
from cal
where cal.Percentage <=100
order by Percentage DESC limit 30;
"""

dft = client.query(top).to_dataframe()
```

```python
plt.figure(figsize=(15,8))
ax = sns.barplot(x='X',y='Percentage',data=dft)
# for i in ax.containers:
#     ax.bar_label(i,)
plt.xticks(rotation = 50)
plt.xlabel('County Code',fontsize=20)
plt.ylabel('Percentage',fontsize=20)
plt.title('2021 County Level: Top 30, Percentage of Deployment Rate',fontsize=30)
plt.show()
plt.savefig('2021 County Level: Top 30, Percentage of Deployment Rate.png')
```

# Bottom 30 including 0

```python
#Bottom 30 
bottom ="""
with cal as(
  SELECT 
Countycode_2 as X,
IFNULL(BroadbandNum/Nums*100,0) as Percentage
FROM `broadband-data.CountyInfo.2021_countyinfo` 
order by Percentage DESC)

select X,
cal.Percentage
from cal
where cal.Percentage <=100
order by Percentage ASC limit 30;
"""
dfb = client.query(bottom).to_dataframe()
```

```python
plt.figure(figsize=(15,8))
ax = sns.barplot(x='X',y='Percentage',data=dfb)
# for i in ax.containers:
#     ax.bar_label(i,)
plt.xticks(rotation = 50)
plt.xlabel('County Code',fontsize=20)
plt.ylabel('Percentage',fontsize=20)
plt.title('2021 County Level: Bottom 30 including 0, Percentage of Deployment Rate',fontsize=30)
plt.show()
plt.savefig('2021 County Level: Bottom 30 including 0, Percentage of Deployment Rate.png')
```

# Bottom 30 exclude 0

```python
# Bottom None 0
bottom_none ="""
with cal as(
  SELECT 
Countycode_2 as X,
IFNULL(BroadbandNum/Nums*100,0) as Percentage
FROM `broadband-data.CountyInfo.2021_countyinfo` 
order by Percentage DESC)

select X,
cal.Percentage
from cal
where cal.Percentage <=100 and Percentage >0
order by Percentage ASC limit 30;
"""
dfbn0 = client.query(bottom_none).to_dataframe()
```
