```python
import pandas as pd

data = pd.read_csv("cc-est2019-alldata.csv")
```

```python
data.columns
```

```python
data_small = data[['STATE', 'COUNTY', 'STNAME', 'CTYNAME', 'YEAR','AGEGRP','TOT_MALE','WA_MALE', 'WA_FEMALE']][data.YEAR ==1]
```

```python
data_few_state = data_small[data_small.STNAME.isin(['Alabama','Arizona','Arkansas','California','Colorado','Connecticut','Virginia','Missouri'])]
data_few_state['White population'] = data_few_state['WA_MALE']+data_few_state['WA_FEMALE']
data_few_state
```

```python
data_few_state_aggregate = data_few_state.groupby(['STNAME','COUNTY','CTYNAME'])[['TOT_MALE','White population']].sum()
```

```python
pd.DataFrame(data_few_state_aggregate).to_csv("data_few_state_aggregate.csv")
```
