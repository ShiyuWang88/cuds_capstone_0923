# Fiber/DSL/Cable Nums

## 2021

```python
fiber2021="""
SELECT StateAbbr as State,
count(*) as Nums
FROM `broadband-data.fcc_form_477.fbd_us_without_satellite_jun2021_v1` 
where TechCode = 50 and Consumer=1
group by State
order by Nums DESC
"""
dsl2021="""
SELECT StateAbbr as State,
count(*) as Nums
FROM `broadband-data.fcc_form_477.fbd_us_without_satellite_jun2021_v1` 
where TechCode in (10,11,12,20) and Consumer = 1
group by State
order by Nums DESC
"""

cable2021="""
SELECT StateAbbr as State,
count(*) as Nums
FROM `broadband-data.fcc_form_477.fbd_us_without_satellite_jun2021_v1` 
where TechCode in (40,41,42,43) and Consumer =1
group by State
order by Nums DESC
"""

fiber2021df = client.query(fiber2021).to_dataframe()
dsl2021df = client.query(dsl2021).to_dataframe()
cable2021df = client.query(cable2021).to_dataframe()
```

```python
#fiber 2021
plt.figure(figsize=(15,8))
sns.barplot(x='State',y='Nums',data=fiber2021df)
plt.xticks(rotation = 45)
plt.xlabel('State',fontsize=20)
plt.ylabel('Counts',fontsize=20)
plt.title('Census Block: Numbers of Fiber 2021',fontsize=30)
plt.tight_layout()
plt.show()
plt.savefig('Census Block: Numbers of Fiber 2021.png')
```

```python
#DSL 2021
plt.figure(figsize=(15,8))
sns.barplot(x='State',y='Nums',data=dsl2021df)
plt.xticks(rotation = 45)
plt.xlabel('State',fontsize=20)
plt.ylabel('Counts',fontsize=20)
plt.title('Census Block: Numbers of DSL 2021',fontsize=30)
plt.tight_layout()
plt.show()
plt.savefig('Census Block: Numbers of DSL 2021.png')
```

```python
#Cable 2021
plt.figure(figsize=(15,8))
sns.barplot(x='State',y='Nums',data=cable2021df)
plt.xticks(rotation = 45)
plt.xlabel('State',fontsize=20)
plt.ylabel('Counts',fontsize=20)
plt.title('Census Block: Numbers of Cable 2021',fontsize=30)
plt.tight_layout()
plt.show()
plt.savefig('Census Block: Numbers of Cable 2021.png')
```

# Deployment over State

## 2021

```python
# 2021 county deployment 
dr2021= """
SELECT
Countycode_2 as X,
Statecode,
b.State_Name as Statename,
IFNULL(BroadbandNum/Nums*100,0) as Percentage
FROM `broadband-data.CountyInfo.county_2021_having_statecode` as a
inner join `fcc_form_477.State_Block_Counts_2010_2020` as b
on a.Statecode=CAST(b.State_Code as String)
order by Statecode ASC,Percentage DESC
"""

df2021dr = client.query(dr2021).to_dataframe()
```

```python
plt.figure(figsize=(15,8))
ax = sns.stripplot(data=df2021dr,y='Statename',x='Percentage')
# for i in ax.containers:
#     ax.bar_label(i,)
#plt.xticks(rotation = 70)
plt.ylabel('State',fontsize=20)
plt.xlabel('Percentage',fontsize=20)
plt.title('2021 BroadBand Deployment Rate of County over State',fontsize=30)
plt.show()
```
